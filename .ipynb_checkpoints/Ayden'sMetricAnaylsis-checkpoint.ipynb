# Best Neighborhood in Pittsburgh

## Introduction

I aim to determine the best neighborhood in Pittsburgh based on police incident blotter data from the past 30 days. I used Python, Pandas, and matplot to analyze the data and come up the best neighborhood.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

## The Metric

Our metric for determining the best neighborhood is simply the neighborhood with the lowest number of police incidents in the past 30 days. We used a dataset of police incident blotter data from the City of Pittsburgh Open Data Portal (https://data.wprdc.org/dataset/police-incident-blotter) to calculate the number of incidents for each neighborhood.

## The Best Neighborhood

```python
# Load the police incidents data
police_data = pd.read_csv('police_incidents.csv')

# Calculate the number of incidents in each neighborhood
neighborhood_counts = police_data['INCIDENTNEIGHBORHOOD'].value_counts()

# Create a bar chart of the neighborhood incident counts
fig, ax = plt.subplots(figsize=(20,10))
ax.bar(neighborhood_counts.index, neighborhood_counts.values)

# Set the chart title and axis labels
ax.set_title('Police Incidents in Pittsburgh Neighborhoods (Last 30 Days)')
ax.set_xlabel('Neighborhood')
ax.set_ylabel('Number of Incidents')


# Rotate the x-axis labels for readability
plt.xticks(rotation=90)

# Show the chart
plt.show()
```

Cool that makes it pretty visually obvious based on the data who the contenders for the best neighborhood are. Let's write code to tell us exactly which

```python
neighborhood_counts = police_data.groupby('INCIDENTNEIGHBORHOOD').size().reset_index(name='count')

# Sort neighborhoods by the number of incidents in ascending order
neighborhood_counts = neighborhood_counts.sort_values('count')

# Get the neighborhood with the lowest number of incidents
best_neighborhood = neighborhood_counts.iloc[0]['INCIDENTNEIGHBORHOOD']
print(best_neighborhood + ' is the best neighborhood!')
```

After analyzing the data, it has been determined that the best neighborhood in Pittsburgh based on this metric is **Chartiers City**
